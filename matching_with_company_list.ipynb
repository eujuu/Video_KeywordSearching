```python
# scripts 불러오기
scripts = []
with open('/content/drive/My Drive/hana_scripts.txt', 'r') as f:
  para = f.readlines()
  for line in para:
    scripts.append(line.rstrip().lstrip())
scr_time = []
with open('/content/drive/My Drive/hanascripts_timestamps_ver2.txt', 'r') as f:
  data = f.readlines()
  for line in data:
   scr_time.append(line.rstrip())
new_scr_time = []
for w in scr_time:
    new_scr_time.append(w.split(', '))
```

```python
#키워드 설정
company_list = ["코덱스레버리지", "etf", "씨아이에스", "씨에스베어링", "ddr5", "에스폴리텍", "진성티이씨", "크린앤사이언스", "서진시스템", "코리아써키트", "아모텍", "켐트로닉스", "심텍", "유니테스트", "테크윙", "ISC", "동진쎄미캠", "이엔에프테크놀로지", "이오테크닉스", "코스모신소재", "나노신소재", "아이에스동서", "와이엔텍"]
```

```python
time_script = ""
time_stamp = ""
for i in new_scr_time:
  time_script+=(i[0]+" ")
  time_stamp+= (i[1] + " ")

print(time_script)
```

```python
print(time_stamp)
```

```python
time_script_arr = time_script.split(" ")
time_stamp_arr = time_stamp.split(" ")
```

```python
#단어 매칭- 첫출현만 저장
results = []
res = dict()
for i in company_list:
  res[i] = []
for i in company_list:
  search_word = i
  for idx, word in enumerate(time_script_arr):
    if search_word == word:
      results.append(int(time_stamp_arr[idx]))
      res[search_word].append(int(time_stamp_arr[idx]))
      print(search_word)
      break
print(results)
```

```python
#정렬
import numpy as np
print(results)
results = np.array(results)
results = np.sort(results)
results
```
