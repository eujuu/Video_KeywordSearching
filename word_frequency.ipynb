```python
from konlpy.tag import Kkma
kkma = Kkma()
scripts = []
with open('/content/drive/My Drive/hana_scripts.txt', 'r') as f:
  para = f.readlines()
  for line in para:
    scripts.append(line.rstrip())
ex_sent = []
for sen in scripts:
  ex_sent.extend( kkma.sentences(sen))
  #ex_sent = kkma.sentences(sen)
ex_sent
```

```python
from re import match
nouns = []
for sent in ex_sent : 
    for noun in kkma.nouns(sent) :
        # 단어 전처리 : 2음절 이상, 수사 제외
        if len(str(noun)) >= 2 and not(match('^[0-9]', noun)) :
            nouns.append(noun)
```

```python
word_count = {}
for noun in nouns :
    word_count[noun] = word_count.get(noun, 0) + 1
word_count
```

```python
from collections import Counter

counter = Counter(word_count)

top5 = counter.most_common(5)
top5
```

```python
scripts_with_timestamp = []
with open('/content/drive/My Drive/hanascripts_timestamps_ver2.txt', 'r') as f: #타임스탬프 데이터 열기
  data = f.readlines()
  for line in data:
   scripts_with_timestamp.append(line.rstrip())

scripts_with_timestamp
```

```python
new_arr = []
for w in scripts_with_timestamp:
    new_arr.append(w.split(', '))
print(new_arr)
```

```python
import numpy as np
time_report = []
for i in range(0, len(top5)):
  freq_word = top5[i][0]
  for j in range(len(new_arr)):
      if new_arr[j][0] == freq_word:
        
        time_report.append(int(new_arr[j][1]))
        #time_report.append(new_arr[j][2])
time_report = np.array(time_report)
#time_report = time_report.astype('uint8')
print(len(time_report))
#time_report = np.reshape(time_report, (-1, 2))
time_report = np.sort(time_report)
time_report
```
